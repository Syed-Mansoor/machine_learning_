# diabetes prediction

```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
```

```python
print(dataset['DESCR'])
```

```python
dataset
```

```python
import pandas as pd
df_diabetes=pd.DataFrame(dataset.data,columns=['age',
  'sex',
  'bmi',
  'bp',
  's1',
  's2',
  's3',
  's4',
  's5',
  's6'])
df_diabetes.head()
```

```python
## Independent and dependent features
X=df_diabetes
y=dataset['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 10)
```

```python
x_train
```

```python
# correlation
x_train.corr()
```

```python
# heatmap
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize = (15,16))

sns.heatmap(x_train.corr(),annot = True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(x_train,y_train)
```

# Hyperparameter tunning

```python
param={
    'criterion':['squared_error','friedman_mse','absolute_error'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,10,15,20,25],
    'max_features':['auto','sqrt','log2']
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV
model  = DecisionTreeRegressor()
grid  = GridSearchCV(estimator=model,param_grid=param,n_jobs=-1,cv = 5,scoring = 'neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
# prediction
y_pred = grid.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
select_model = DecisionTreeRegressor(criterion='squared_error',max_depth=5,max_features='sqrt',splitter='random')
select_model.fit(x_train,y_train)
```

```python
# visualisse the decision tree
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Plot Decision Tree
plt.figure(figsize=(12, 8))
plot_tree(select_model, filled=True ,rounded=True)
plt.show()
```
