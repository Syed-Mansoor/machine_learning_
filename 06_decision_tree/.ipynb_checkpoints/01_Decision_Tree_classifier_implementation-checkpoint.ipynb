# Decision Tree classifier Implementation

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
iris = load_iris()
```

```python
iris
```

```python
iris.DESCR
```

```python
iris.target
```

```python
##independent features
x=pd.DataFrame(iris['data'],columns=['sepal length in cm','sepal width','petal length','petal width'])
```

```python
# dependent features
y = iris['target']
```

```python
x
```

```python
y
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 10)
```

```python
x_train
```

```python
# apply deicision tree classifier
from sklearn.tree import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(x_train,y_train
        )
```

```python
# visualisse the decision tree
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Plot Decision Tree
plt.figure(figsize=(12, 8))
plot_tree(treeclassifier, filled=True, feature_names=iris.feature_names, class_names=iris.target_names, rounded=True)
plt.show()
```

# hyperparameter Tunning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
import numpy as np

param_dist = {
    'max_depth': [None] + list(np.arange(1, 110, 10)),
    'min_samples_split': np.arange(2, 20, 2),
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'min_samples_leaf': np.arange(1, 20, 2),
    'max_features': ['auto',None, 'sqrt', 'log2']
}

grid = GridSearchCV(DecisionTreeClassifier(),param_grid=param_dist, cv=5,scoring = 'accuracy',n_jobs  = -1)
grid.fit(x_train, y_train)
print(grid.best_params_)
```

```python
y_pred = grid.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python

from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
```

```python
print(f'{score:}')
```

```python
select_model = DecisionTreeClassifier(criterion='gini', max_depth=101, max_features= None, min_samples_leaf= 7, min_samples_split= 16, splitter='random')
select_model.fit(x_train,y_train)
```

```python
# visualisse the decision tree
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Plot Decision Tree
plt.figure(figsize=(12, 8))
plot_tree(select_model, filled=True,rounded=True)
plt.show()
```
