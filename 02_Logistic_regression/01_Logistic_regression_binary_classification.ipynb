# Logistic Regression implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
```

```python
## create the dataset
X, y = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=15)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.30,random_state  = 42)
```

```python
x_train
```

```python
print(x_train.shape,x_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
regression = LogisticRegression()
regression.fit(x_train,y_train)
y_pred = regression.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

# HyperParameter Tunning and cross validation

```python
model = LogisticRegression()
penality = ['l1','l2','elasticnet']
c_values = [100,1,10,2,0.1,0.01]
solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
```

```python
params = dict(penalty =penality,C=c_values,solver =solver)
```

```python
# Grid Search CV
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
grid = GridSearchCV(estimator=model,param_grid=params,scoring = 'accuracy',cv = cv,n_jobs = -1,)
```

```python
grid
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(x_test)
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

```python
# Randomized search cv
from sklearn.model_selection import RandomizedSearchCV
random = RandomizedSearchCV(estimator=model,param_distributions=params,cv= 5,scoring = 'accuracy',n_jobs=-1)
random.fit(x_train,y_train)
```

```python
random.best_estimator_
```

```python
random.best_params_
```

```python
random.predict(x_test)
```
