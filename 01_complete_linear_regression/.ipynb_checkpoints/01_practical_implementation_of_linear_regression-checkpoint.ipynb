```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('height-weight.xls')
df.head()
```

```python
sns.scatterplot(data  =df,x=df['Weight'],y=df['Height'])
plt.xlabel('Weight')
plt.ylabel('Height')
plt.show()
```

```python
# correlation
df.corr()
```

```python
# seaborn for visualisation
sns.heatmap(df)
```

```python
sns.pairplot(df)
```

```python
# independent and dependent features
from sklearn.model_selection import train_test_split
x = df[['Weight']] # indepedent feautres should be data frame or 2 dimensional array
y = df['Height'] # this variable can be in 1 dimension series or 1d array
```

```python
x.head()
```

```python
y.head()
```

```python
# train test split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state =42)
```

```python
print(x_train.shape,x_test.shape)
```

```python
# standarization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train= scaler.fit_transform(x_train)
x_test= scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
# apply linear regression
from sklearn.linear_model import LinearRegression
regression = LinearRegression(n_jobs=-1)
```

```python
regression.fit(x_train,y_train)
```

```python
y_pred = regression.predict(x_test)
```

```python
y_pred
```

```python
print('Coeffient or slope :',regression.coef_ ) # coeffient
```

```python
print('Intercept :',regression.intercept_ )# intercept
```

```python
# plot best fit line
plt.scatter(x_train,y_train)
plt.plot(x_train,regression.predict(x_train),color = 'red')
```

```python
y_pred # prediction for test data
```

```python
# performance metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error

mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)

rmse = np.sqrt(mse)
print(f'Mean absolute error : {mae}')
print(f'Mean Squarred error :{mse}')
print(f'Root mean squarred error :{rmse}')
```

# R Square
Formula

R ^ 2 = 1- SSR/SST
R ^ 2 = coeffient of determination SSR = sum of squarres of residuals SST = total sum of squarres

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,y_pred)
print(f'{score : }')
```

# Adjusted R2 Score

r2 = 1 - [(1 - R2) * (N -1)/(N - k -1)]

where :
r2: the r2 of the model
n: the number of observations 
k: the number of predictor variables

```python
# display adjusted r_squarred
1 - (1 - score) * (len(y_test) - 1) / (len(y_test) - x_test.shape[1] -1 )
```
