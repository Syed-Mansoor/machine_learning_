```python
df = sns.load_dataset('tips')
```

```python
# independent and dependent features
x = df.drop(columns = ['total_bill'])
y = df['total_bill']
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state = 10)
```

```python
print(x_train.shape,x_test.shape)
```

# feature encoding
- . One hot encoding
- . Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()

x_train['sex'] = le1.fit_transform(x_train['sex'])
x_train['time'] = le2.fit_transform(x_train['time'])
x_train['smoker'] = le3.fit_transform(x_train['smoker'])



x_test['sex'] = le1.transform(x_test['sex'])


x_test['time'] = le2.transform(x_test['time'])
x_test['smoker'] = le3.transform(x_test['smoker'])
```

```python
x_train.head()
```

```python
# one hot encoding
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct  = ColumnTransformer(transformers=[('onehot',OneHotEncoder(drop = 'first'),[3])],remainder='passthrough')



import sys
import numpy
numpy.set_printoptions(threshold=sys.maxsize)
x_train = ct.fit_transform(x_train)
x_test = ct.transform(x_test)
```

```python
# support vector regressor
from sklearn.svm import SVR
svr = SVR()
```

```python
svr.fit(x_train,y_train)
```

```python
y_pred = svr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

# Hyper parameter tunning

```python
from sklearn.model_selection import GridSearchCV
 
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
```

```python
grid = GridSearchCV(SVR(), param_grid, refit = True, verbose = 3)
 
# fitting the model for grid search
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
grid_prediction=grid.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
print(r2_score(y_test,grid_prediction))
print(mean_absolute_error(y_test,grid_prediction))
```
