
# Naive Baye's Machine Learning Algorithms Implementation

```python
from sklearn.datasets import load_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
X,y=load_iris(return_X_y=True)
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
x_train
```

```python
from sklearn.naive_bayes import GaussianNB
gnb=  GaussianNB()
gnb.fit(x_train,y_train)
```

```python
y_pred =gnb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
import seaborn as sns
```

```python
df  = sns.load_dataset('tips')
```

```python
df.head(

)
```

```python
df.iloc[:,:-2:]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop(columns = 'time'),df['time'],test_size = 0.25,random_state = 42)
```

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()

x_train['sex'] = le1.fit_transform(x_train['sex'])
# x_train['time'] = le2.fit_transform(x_train['time'])
x_train['smoker'] = le3.fit_transform(x_train['smoker'])



x_test['sex'] = le1.transform(x_test['sex'])


# x_test['time'] = le2.transform(x_test['time'])
x_test['smoker'] = le3.transform(x_test['smoker'])
```

```python
x_train
```

```python
# one hot encoding
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct  = ColumnTransformer(transformers=[('onehot',OneHotEncoder(drop = 'first'),[4])],remainder='passthrough')



import sys
import numpy
numpy.set_printoptions(threshold=sys.maxsize)
x_train = ct.fit_transform(x_train)
x_test = ct.transform(x_test)
```

```python
x_train
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,CategoricalNB,ComplementNB
gnb=  GaussianNB()
gnb.fit(x_train,y_train)
```

```python
y_pred1 = gnb.predict(x_test)
print(confusion_matrix(y_pred1,y_test))
print(accuracy_score(y_pred1,y_test))
print(classification_report(y_pred1,y_test))
```

```python
mnb= MultinomialNB()
mnb.fit(x_train,y_train)
y_pred2 = mnb.predict(x_test)
print(confusion_matrix(y_pred2,y_test))
print(accuracy_score(y_pred2,y_test))
print(classification_report(y_pred2,y_test))
```

```python
cnb= CategoricalNB()
cnb.fit(x_train,y_train)
y_pred3 = cnb.predict(x_test)
print(confusion_matrix(y_pred3,y_test))
print(accuracy_score(y_pred3,y_test))
print(classification_report(y_pred3,y_test))
```

```python
ccnb= ComplementNB()
ccnb.fit(x_train,y_train)
y_pred4 = ccnb.predict(x_test)
print(confusion_matrix(y_pred4,y_test))
print(accuracy_score(y_pred4,y_test))
print(classification_report(y_pred4,y_test))
```
